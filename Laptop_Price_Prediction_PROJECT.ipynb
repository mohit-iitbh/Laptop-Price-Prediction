```python
## Defining a function for fetching the different processor names from CPU_Name column
def fetch_processor_names(data):
  if data == 'Intel Core i7' or data == 'Intel Core i5' or data == 'Intel Core i3' or data == 'Intel Celeron Dual' or data == 'Intel Pentium Quad' or data == 'Intel Core M':
    return data
  else:
    if data.split()[0] == 'Intel':
      return 'Other Intel Processor'
    else:
      return 'AMD Processor'
```

```python
# We are creating four new columns which will store various information present in "Memory".
# One of the columns is for SSD value, the next one stores HardDrive, the next one is Flash_Storage and the last one is Hybrid.
df['Memory'] = df['Memory'].astype(str).replace('\.0','',regex=True)
df['Memory'] = df['Memory'].astype(str).replace('GB','')
df['Memory'] = df['Memory'].astype(str).replace('TB','000')
n = df['Memory'].str.split("+",expand = True)
```

```python
df['first'] = n[0]
df['first'] = df['first'].str.strip()
```

```python
df["2nd"] =n[1]
```

```python
# Creating temporary columns to store values
df['Layer1HDD'] = df['first'].apply(lambda x:1 if "HDD" in x else 0)
df['Layer1SSD'] = df['first'].apply(lambda x:1 if "SSD" in x else 0)
df['Layer1Hybrid'] = df['first'].apply(lambda x:1 if "Hybrid" in x else 0)
df['Layer1Flash_Strorage'] = df['first'].apply(lambda x:1 if "Flash Storage" in x else 0)
```

```python
df['first'] = df['first'].str.replace(r'\D','')
df['2nd'].fillna("0",inplace = True)
df["Layer2HDD"] = df['2nd'].apply(lambda x:1 if 'HDD' in x else 0)
df["Layer2SSD"] = df['2nd'].apply(lambda x:1 if 'SSD' in x else 0)
df["Layer2Hybrid"] = df['2nd'].apply(lambda x:1 if 'Hybrid' in x else 0)
df["Layer2Flash_Storage"] = df['2nd'].apply(lambda x:1 if 'Flash Storage' in x else 0)

```

```python
df['2nd'] =df['2nd'].str.replace(r'\D','')
# Here we changing the data type of column first and 2nd into integer
df['first'] = df['first'].astype(int)
df['2nd'] = df['2nd'].astype(int)


```

```python
df["HardDrive"] =(df['first']*df['Layer1HDD']+df['2nd']*df['Layer2HDD'])
df["SSD"] =(df['first']*df['Layer1SSD']+df['2nd']*df['Layer2SSD'])
df["Hybrid"] =(df['first']*df['Layer1Hybrid']+df['2nd']*df['Layer2Hybrid'])
df["FlashStorage"] =(df['first']*df['Layer1Flash_Strorage']+df['2nd']*df["Layer2Flash_Storage"])
```

```python
# Dropping all temporary column which is not useful anymore
df.drop(columns =['first','2nd','Layer1Hybrid','Layer2Hybrid','Layer1HDD','Layer2HDD','Layer1SSD','Layer2SSD','Layer1Flash_Strorage',"Layer2Flash_Storage"],inplace = True)
```

# **Spliting data into Training_data and Testing_data**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state = 2)
```

```python
x_train
```

# **Building and Training the Model**

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

### **Linear Regression**

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(x_train,y_train)

y_pred = pipe.predict(x_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

**RESULT:**
- we achieved **79.06%** accuracy with Linear Regression Model

### **Ridge Regression**

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = Ridge(alpha=3)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(x_train,y_train)

y_pred = pipe.predict(x_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

**RESULT:**
- With Ridge Regression we got an accuracy of **79.33%** by tuning the hyperparameter: alpha

### **Decision Tree Regressor**

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = DecisionTreeRegressor(max_depth=10, splitter = 'random')

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(x_train,y_train)

y_pred = pipe.predict(x_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

**RESULT:**
- With Decision Tree Regressor we got an accuracy of **80.11%** by tuning the hyperparameters: max_depth and splitter.

### **Random Forest Regressor**

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=13)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(x_train,y_train)

y_pred = pipe.predict(x_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

**RESULT:**
- With Random Forest Regressor we got an accuracy of **85.36%** by tuning the hyperparameters.

### **XG Boost Regressor**

```python
# import joblib

step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = XGBRegressor(n_estimators=40,max_depth=6,learning_rate=0.5)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(x_train,y_train)

y_pred = pipe.predict(x_test)

# model_filename = 'xgboost_model.pkl'
# joblib.dump(pipe, model_filename)


print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

**RESULT:**
- With XG Boost Regressor we got highest accuracy of **86.64%** by tuning the hyperparameters.
